# Importing Libraries

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
# conda install -c conda-forge mlxtend
```

```python
titanic = pd.read_csv("Titanic.csv")
titanic.head()
```

# Pre Processing

```python
df=pd.get_dummies(titanic)
df.head()
```

# Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)  #we choose the thresold values based on data set if dataset is big thresold will be high
rules
rules.sort_values('lift',ascending = False)
```

An leverage value of 0 indicates independence. Range will be [-1 1]
## A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending = False)[0:20]   #showing for 20 rows 
```

```python
rules[rules.lift>1]
```
